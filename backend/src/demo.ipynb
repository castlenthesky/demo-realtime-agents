```python
import uvicorn
import threading
import asyncio
import time

from src.main import sio_app, sio
from src.config import settings
```

## Launch the socket.io server

```python
# Launch the server


config = uvicorn.Config(sio_app, host="0.0.0.0", port=settings.API_PORT, reload=False)
server = uvicorn.Server(config)

# Warp server in a background thread so it doesn't block the notebook
def run_server():
    loop = asyncio.new_event_loop()
    asyncio.set_event_loop(loop)
    loop.run_until_complete(server.serve())

thread = threading.Thread(target=run_server, daemon=True)
thread.start()

# Give the server a moment to start
time.sleep(1)

print(f"Server starting on http://0.0.0.0:{settings.API_PORT}")
print(f"Socket.IO endpoint: http://0.0.0.0:{settings.API_PORT}/socket.io/")
print(f"Server thread running: {thread.is_alive()}")
```

```python
# Emit a message to all connected clients
await sio.emit("message", {"message": "Hello from the server!"})
```
